```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
def freq_analysis(V, HR, Fs):
    n = len(V)

    window = signal.windows.hann(n)
    V_win = V * window
    HR_win = HR * window

    fft_V = np.fft.rfft(V_win) / n
    fft_HR = np.fft.rfft(HR_win) / n
    freq = np.fft.rfftfreq(n, Fs)

    power_V = np.abs(fft_V)**2 * (n / np.sum(window**2))
    cross_power = fft_HR * np.conj(fft_V) * (n / np.sum(window**2))

    H = cross_power / (power_V + np.finfo(float).eps)
    amplitude = np.abs(H)
    phase_deg = np.rad2deg(np.angle(H))

    max_index = np.where(freq <= 0.3)[0][-1]
    freq = freq[:max_index+1]
    amplitude = amplitude[:max_index+1]
    phase_deg = phase_deg[:max_index+1]

    return amplitude, phase_deg, freq
```

```python
Ts = 0.01
Fs = 1 / Ts
max_time = 200
```

```python
Kp = 6
Ks = 18
Ap = 2.5
As = 0.4
fp = 0.2
fs = 0.015

t = np.arange(0, 300, Ts)
V_normal = 0.3 * signal.chirp(t, 0.005, 300, 0.5, method='linear')
HR_normal = 5 * V_normal + np.random.randn(len(V_normal)) * 2
ABP_normal = 0.01 * np.cumsum(HR_normal) * Ts + 0.3 * V_normal

time_model = t

fig, axs = plt.subplots(3, 1, figsize=(15, 10))
axs[0].plot(time_model[:int(max_time/Ts)], V_normal[:int(max_time/Ts)], color='blue')
axs[1].plot(time_model[:int(max_time/Ts)], HR_normal[:int(max_time/Ts)], color='red')
axs[2].plot(time_model[:int(max_time/Ts)], ABP_normal[:int(max_time/Ts)], color='green')
plt.suptitle("Физиологические показатели")
plt.tight_layout()
plt.show()

amplitude_normal, phase_deg_normal, freq_normal = freq_analysis(V_normal, HR_normal, Fs)

fig, axs = plt.subplots(2, 1, figsize=(15, 10))
axs[0].plot(freq_normal, amplitude_normal, color='blue', lw=2.5)
axs[1].plot(freq_normal, phase_deg_normal, color='red', lw=2.5)
axs[1].set_ylim(-180, 180)
plt.suptitle("Анализ передаточной функции: V → HR")
plt.tight_layout()
plt.show()
```
