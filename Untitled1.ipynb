```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import chirp
from scipy.signal import cont2discrete
from scipy.signal import lfilter
from scipy.signal.windows import hann
from scipy.fft import rfft, rfftfreq

plt.rcParams["figure.figsize"] = (12,6)
plt.rcParams["font.size"] = 12
```

# параметры объекта

```python
Ts = 0.01
Fs = 1 / Ts

time_simulation = 300

time = np.arange(0, time_simulation, Ts)

# Параметры дыхания
freq_initial = 0.005
freq_target = 0.5

# Задержки

time_delay_sym = 2
time_delay_vas = 0.42
time_delay_bar = 0.30

# Параметры модели

Kp = 6
Ks = 18

Ap = 2.5
As = 0.4

fp = 0.2
fs = 0.015
```

```python
V = 0.3 * chirp(
    time,
    f0=freq_initial,
    f1=freq_target,
    t1=time_simulation,
    method="linear"
)

plt.figure(figsize=(15,4))
plt.plot(time, V)
plt.xlim(0,200)
plt.grid(True)
plt.xlabel("Время, с")
plt.ylabel("Объем легких")
plt.title("Дыхательный сигнал")
plt.show()
```

```python
def delay(signal, delay_sec, Ts):

    n = int(delay_sec/Ts)

    if n == 0:
        return signal.copy()

    out = np.zeros_like(signal)

    out[n:] = signal[:-n]

    return out
```

```python
def first_order_filter(signal, K, fc, Ts):

    wc = 2*np.pi*fc

    alpha = wc*Ts/(1+wc*Ts)

    y = np.zeros_like(signal)

    for i in range(1,len(signal)):
        y[i] = y[i-1] + alpha*(K*signal[i]-y[i-1])

    return y
```

```python
def freq_analysis(V, HR, Fs):

    n = len(V)

    window = hann(n)

    Vw = V*window
    HRw = HR*window

    FFT_V = rfft(Vw)/n
    FFT_HR = rfft(HRw)/n

    freq = rfftfreq(n,1/Fs)

    power = np.abs(FFT_V)**2

    cross = FFT_HR*np.conj(FFT_V)

    H = cross/(power+1e-12)

    amplitude = np.abs(H)

    phase = np.degrees(np.angle(H))

    mask = freq<=0.3

    return amplitude[mask], phase[mask], freq[mask]
```

```python
from scipy.signal import cont2discrete, lfilter

def lowpass(num, den, Ts):

    b, a, _ = cont2discrete((num, den), Ts, method='zoh')[:3]

    b = b.flatten()

    a = a.flatten()

    return b, a


# Блуждающая ветвь

num_vagal = [-Kp * 2 * np.pi * fp]
den_vagal = [1, 2 * np.pi * fp]

b_vagal, a_vagal = lowpass(
    num_vagal,
    den_vagal,
    Ts
)

# Симпатическая ветвь

num_sym = [Ks * 2 * np.pi * fs]
den_sym = [1, 2 * np.pi * fs]

b_sym, a_sym = lowpass(
    num_sym,
    den_sym,
    Ts
)
```

```python
def simulate(Kp, Ks, Ap, As, fp, fs):

    V = 0.3 * chirp(
        time,
        f0=freq_initial,
        f1=freq_target,
        t1=time_simulation,
        method="linear"
    )

    # Передаточные функции

    num_v = [-Kp * 2 * np.pi * fp]
    den_v = [1, 2 * np.pi * fp]

    b_v, a_v = lowpass(num_v, den_v, Ts)

    num_s = [Ks * 2 * np.pi * fs]
    den_s = [1, 2 * np.pi * fs]

    b_s, a_s = lowpass(num_s, den_s, Ts)

    ABP = np.zeros(len(V))
    HR = np.zeros(len(V))

    mech = -np.gradient(V, Ts)

    for _ in range(8):

        bar = 0.01 * delay(ABP, time_delay_bar, Ts)

        vagal = Ap * (bar - V)

        sym = As * (-bar - V)

        vagal = lfilter(b_v, a_v, vagal)

        sym = delay(sym, time_delay_sym, Ts)

        sym = lfilter(b_s, a_s, sym)

        HR = vagal + sym

        vessel = 0.01 * delay(HR, time_delay_vas, Ts)

        ABP = vessel + mech

    return V, HR, ABP
```

# Моделирование

```python
V_normal, HR_normal, ABP_normal = simulate(
    Kp,
    Ks,
    Ap,
    As,
    fp,
    fs
)
```

```python
fig, ax = plt.subplots(3,1, figsize=(15,10))

ax[0].plot(time, V_normal)

ax[0].set_title("Изменение объема легких")

ax[0].set_ylabel("V")

ax[0].grid()

ax[1].plot(time, HR_normal, color="red")

ax[1].set_title("Частота сердечных сокращений")

ax[1].set_ylabel("HR")

ax[1].grid()

ax[2].plot(time, ABP_normal, color="green")

ax[2].set_title("Артериальное давление")

ax[2].set_ylabel("ABP")

ax[2].set_xlabel("Время, с")

ax[2].grid()

for a in ax:
    a.set_xlim(0,200)

plt.tight_layout()

plt.show()
```

# Частотный анализ

```python
# Частотный анализ

amplitude_normal, phase_normal, freq_normal = freq_analysis(
    V_normal,
    HR_normal,
    Fs
)

print("Расчет завершен.")
```

```python
plt.figure(figsize=(12,5))

plt.plot(
    freq_normal,
    amplitude_normal,
    linewidth=2,
    color="blue"
)

plt.title("Амплитудно-частотная характеристика")

plt.xlabel("Частота, Гц")

plt.ylabel("|H|")

plt.xlim(0,0.3)

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.plot(
    freq_normal,
    phase_normal,
    linewidth=2,
    color="red"
)

plt.title("Фазочастотная характеристика")

plt.xlabel("Частота, Гц")

plt.ylabel("Фаза, град")

plt.xlim(0,0.3)

plt.ylim(-180,180)

plt.grid(True)

plt.show()
```
